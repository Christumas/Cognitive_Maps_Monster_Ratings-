```python
import pandas as pd
import numpy as np
from PIL import Image
import itertools
import os
import random
import json
```

```python
Stimdirectory = 'C:\JS Psych Experiments\Monster Ratings\monsters'
Stimuli = [filename for filename in os.listdir(Stimdirectory) if filename.endswith(('.jpeg', '.png')) & filename.startswith('obj')] #get all the stimuli file names
#find all possible unique combinations of stimuli
uniquepairings = list(itertools.combinations(Stimuli,2)) #this list of unique pairings then can be shuffled for each participant
```

```python
#Check if the correct stimuli are included
print(Stimuli)
print(len(Stimuli))
```

```python
DesignFile = {'Participant':[],
              'Block':[],
              'Trial':[],
              'StimLeft':[],
              'StimRight':[],
} #basic structure of the design file

Dataframe = pd.DataFrame(DesignFile) 
Dataframe

#lets create a design file generator
NumParticipants =  100
NumTrials = len(uniquepairings)
NumBlocks = 1

for participant in range(NumParticipants+1): #specify the number of iterations (design files) based off the number of participants
    #print(participant)
    Dataframe['Participant'] = np.repeat(participant+1, NumTrials)
    Dataframe['Block'] = np.repeat(NumBlocks, NumTrials)
    Dataframe['Trial'] = range(NumTrials) 
    Dataframe['Trial'] += 1
    
    shuffledPairs = [] #variable to hold the set of shuffled unique pairings
    #lets shuffle the contents of the pairs themselves
    for i in uniquepairings: #loop through all the unique pairings
        pair = list(i) #convert it to list so we can use shuffle
        random.shuffle(pair) #we shuffle the contents inside each pair
        shuffledPairs.append(tuple(pair)) 
    
    random.shuffle(shuffledPairs) #then shuffle the entire list of unique pairs
    
    #assign the left and right stimulus to be shown
    Dataframe['StimLeft'] = [i[0] for i in shuffledPairs]
    Dataframe['StimRight'] = [i[1] for i in shuffledPairs]
    
    ppnumber = str(participant + 1).zfill(3);
    
    Dataframe.to_csv(f'DesignFiles/PP{ppnumber}.csv', index=False)

        
    
    

Dataframe
```

```python
#create a json file for Jatos with the list of participant design files for Jatos
designpath= 'DesignFiles'
designfiles = [file for file in os.listdir(designpath) if file.endswith('.csv')]

jsonPath = os.path.join(designpath, "designfiles.json")

#save json file
with open(jsonPath, "w") as json_file:
    json.dump(designfiles, json_file, indent=4)

```
